```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data_path = "Lesson_4_student_grades_dataset_3.csv"
dataset = pd.read_csv(data_path)

statistics = dataset.describe()
a = dataset['Hours_of_Study']
b = dataset['Attendance_Percentage']
c = dataset['Previous_Exam_Scores']
d = dataset['Extracurricular_Activities']
e = dataset ['Travel_Time_to_School']
f = dataset['Grades']

print(dataset.isnull().sum())
```

```python
plt.figure(facecolor='pink')
plt.scatter(a,f)
plt.xlabel('Hours of Study')
plt.ylabel('Grades')
plt.title('Student Grades Data Set')
```

```python
plt.figure(facecolor='pink')
plt.scatter(b,e)
plt.xlabel('Attendance Percentage')
plt.ylabel('Travel Time to School')
plt.title('Student Grades Data Set')
```

```python
plt.figure(facecolor='pink')
plt.scatter(c,d)
plt.xlabel('Previous Exam Scores')
plt.ylabel('Extracurricular Activities')
plt.title('Student Grades Data Set')
```

Prepare for Data Modeling

```python
x = dataset.drop('Grades', axis=1)
y = dataset['Grades']
```

Split and Fit Linear Regression

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state = 42)
model = LinearRegression()
model.fit(x_train, y_train)
```

Making Predicitions

```python
new_data = np.array([[6, 91, 87, 1, 50]])
prediction = model.predict(new_data)
print(f"Prediction Grades: {prediction[0]}")
```
